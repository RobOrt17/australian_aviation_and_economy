Reading in my second data set, on economical figures.

```python
economy = pd.read_csv('../australian_aviation_and_economy/data/gdp_percapita.csv')
```

```python
economy.shape
```

```python
economy.head()
```

```python
economy = economy.drop(columns=['Value Footnotes'])
economy
```

vv Changed the names of the columns

```python
economy = economy.rename(columns={'Value': 'GDP Per Capita - Foreign', 'Country or Area': 'Country'})
economy
```

```python
economy.dtypes
```

```python
aus_economy = pd.read_csv('../australian_aviation_and_economy/data/AUS_GDP_per_capita.csv')
aus_economy
```

```python
aus_economy = aus_economy.rename(columns={'GDP per capita (current US$)': 'GDP Per Capita - Domestic'})
aus_economy.head()
```

```python
aus_economy.dtypes
```

```python
aviation.dtypes
```

^^ The 'Year' data type needs to match, in order to merge.

```python
economy['Year'] = economy['Year'].astype(str).astype(int)
economy.dtypes
```

For this dataframe (economy), I will need to extract the GDP per capita($USD), for the the range of years: 1990-2018. This needs to be merged with the aviation dataframe. ^^

```python
aviation_economy = aviation.merge(economy, how='left', on=['Country','Year'], sort=False)
aviation_economy
```

```python
aviation_economy = aviation_economy.merge(aus_economy, how='left', on=['Year'], sort=False)
aviation_economy
```

^^ The above shows that I was able to merge two different economical dataframes on to the aviation dataframe
